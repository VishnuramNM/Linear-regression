```python
teams=teams.dropna()
teams
```

# 3.creating X matrix

```python
X=teams[["athletes","prev_medals"]].copy()
X
```

# 4.creating y matrix

```python
y=teams[["medals"]].copy()
y
```

# 5.adding an extra column of 1 in X as intercept

```python
X["intercept"]=1
```

```python
X=X[["intercept","athletes","prev_medals"]]
```

```python
X
```

# 6.finding  transpose of matrix X

```python
X_T=X.T
```

```python
X_T
```

# 7.find the value of inverse((X_T)X)

```python
M=np.linalg.inv(X_T@X)
M
```

# 8.finding B 

```python
B=M@X_T@y
```

```python
B
```

0-b0-y intercept
1-b1-coefficiant of athletes
2-b2-coeff of prev_medals

```python
B.index=X.columns
```

# 9.Predicted medals,y=XB

```python
predictions=X@B
```
